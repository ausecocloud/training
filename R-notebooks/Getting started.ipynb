## Jupyter Notebooks <a name="notebook"></a>

Jupyter notebooks allow you to create a live document in which you can edit text, code and equations, or share results. The advantage of using a Jupyter notebook is that you are able to work and annotate your data, which you may want to share with colleagues, or use in published materials. 

In *ecocloud*, we use Jupyter notebooks for this reason. It lets you work on a document and save it for later. For more of an introduction on using Jupyter itself, head to our **[support page](https://support.ecocloud.org.au/support/solutions/articles/6000200389-using-jupyter-notebooks)** or on the **[Jupyter website](http://jupyter.org/)**.

Notebooks have three different types of cells:
- Markdown: these are used for comments. Check this page for useful [Markdown tips](https://www.ibm.com/support/knowledgecenter/SSQNUZ_current/com.ibm.icpdata.doc/dsx/markd-jupyter.html).
- Code: these are used for R or Python code. The coding language that you can run depends on what kind of server you have running in *ecocloud*.
- Raw: can be used to write output directly.

Use the commands on the top of the notebook to (from left to right):
- Save your notebook.
- Add a new cell (these are by default code, so if you want the cell to be markdown, change this in the drop-down menu.
- Cut selected cells.
- Copy selected cells.
- Paste from clipboard.
- Execute a cell. 
- Interupt the kernel, this will stop execution of cells.
- Restart the kernel.

## Get started in your *ecocloud* workspace <a name="upload"></a>

When you open your server in *ecocloud* you will automatically see your workspace loaded in the menu on the left. You can open a notebook from here, or find a notebook elsewhere. For example:
- Go to a **Google Drive** folder by clicking the Google Drive icon.
- Go to a **GitHub repository**. As an example: click the GitHub icon (the cat icon, 3rd from the top). By default you will see the *ecocloud* training repository. If you are running an R server, click on the R-notebooks folder and open the notebook you want to work with.

You can also start a new notebook by clicking "File" > "New" > "Notebook".

**Note** If you use a notebook from a GitHub repository, please save it first to your own workspace so any of your changes will be saved. To do this, go to "File" > "Save Notebook As" and use the following: "workspace/my-notebook.ipynb". You can change 'my-notebook' to a more appropriate name. Without the "workspace/" in the name the notebook will not be saved in the workspace for future use. 

If you would like to organise your notebooks in folders you will first need to add a new folder (use the folder icon with the + in the menu bar on the left) before you can save a notebook to that. Make sure you update the "Save Notebook As" to include the new folder name, e.g. "workspace/new-folder/my-notebook.ipynb".

**Adding Data**

Before adding data, make sure you are currently sitting in the 'workspace' folder (as opposed to 'home') or any other folder if you have added these to organise your files and notebooks.

Add data to your workspace by using the upload button (arrow up) and select the file from your computer. It will immediately appear in your workspace items on the left.
For this notebook, you can use the EcolArchives-E089-51-D1.csv file.

You can also upload it through the *[ecocloud Workspace](https://app.ecocloud.org.au/workspace)*, but this will only upload if you don't have a server running!

**Work through an existing notebook**

If you want to work through a notebook without seeing the outputs of pre-run script, go to "Edit" > "Clear All Outputs". This will give you a clean notebook with just the markdown and code cells.

You can execute cells of code in a notebook by clicking on the cell and then press "Shift+Enter" on your keyboard, or click the "Play" icon in the menubar. This will execute the current cell, show any output, and jump to the next cell below. During execution of a cell you will see a * next to the cell.


## An ecological example <a name="example"></a>

Here we use a publically available dataset to show you how to import data in _ecocloud_, and generate summary plots and analyses. This dataset explores the relationship(s) between marine predator and prey body mass. The dataset is taken from the Ecological Archives of the ESA (_Barnes et al_. 2008, Ecology 89:881).

```r
DF <- read.csv("EcolArchives-E089-51-D1.csv")
head(DF)
```

In the first line of code above, we have read the dataset in and named it 'DF'. In the second line we have called the first 6 lines of the dataset into view. An important first step when you load data in is to check the structure and format of the dataset. This may flag whether or not there are issues within the dataset (such as having a letter in a column which should only be numbers). To do this use the following call `str(DF)`. From this, we can see that we have a mix of numbers, factors and integers.

```r
str(DF)
```

## Visualising data in your notebook

There are a several basic plotting commands in R that are relatively straight forward to use. These commands will let you plot summary figures and can help explore some of the data you have. 

### Scatter plots

First we will start by plotting the predator mass against the prey mass to see if there is a relationship between these two variables. Scatterplots are used a lot to represent the relationship between two variables along two axes.

```r
plot(DF$Predator.mass, DF$Prey.mass, pch=16) 
#In R, the variable is called by using dataset$variable-name, so here DF is the dataset, and Predator.mass and Prey.mass are the variables.
```

At a glance, there does not appear to be a strong relationship, however it is likely that we need to transform this dataset. In general small predators are more likely to feed on small prey, and large predators are more likely to feed on larger prey. As such a log transformation may be most appropriate.

```r
plot(log(DF$Predator.mass), log(DF$Prey.mass), pch=16)
```

Now it appears that there is a relationship between the two variables, and even potentially two clusters. 

### Histograms

Another plot that can be generated is a histogram. Histograms are useful to visualise frequency statistics. Here we maintain the log transformation in the histogram plot describing the variation in predator mass. Each bar represents the number of predators with that logged weight. 

We have also included code in the second line to label the x and y axis of the plot.

```r
hist(log(DF$Predator.mass), #which variable to plot
    xlab = "Predator Mass (kg)", ylab = "Count") #labels of axes
```

### Boxplots

A boxplot is another simple way to represent data. Typically the box represents the second and third quartile of the data, the line in the middle of the box represents the median value, and the top and bottom of the box represent the first and fourth quartile, respectively. Any outliers are represented as dots. 

For this dataset, a boxplot can be used to show the distribution of reported predator mass in different locations. We have added another line of code to change the title of the graph.

```r
boxplot(log(DF$Predator.mass) ~ DF$Location,  #which variables to plot
    xlab = "Location", ylab = "Predator Mass", #labels of axes
    main = "Predator mass by location") #title of plot
```
