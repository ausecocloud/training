This example takes a URL to a CSV file and reads it into a variable in python for further processing via the `Requests` library. There is also example of previewing the CSV using the `Pandas` library via Dataframe. `Pandas` is useful for processing this kind of information. Lastly, the notebook shows you how to use `matplotlib` to plot the data frame from the CSV into a simple bar chart.

```python
# specify the URL of the CSV file
url = "https://data.gov.au/dataset/22da2db8-b68f-4853-afb3-964239600d56/resource/b10ea060-2025-4a2d-8171-736de0875e4d/download/2016-soe-atmosphere-average-max-4-hr-average-ozone-concentrations-for-major-cities-1999-2014.csv"

# use Python requests library to read the content to a variable 'csv_content'
import requests
csv_content = None
try:
   r = requests.get(url)
   if(r.status_code != 200):
        raise Exception('')
   csv_content = r.text
except Exception:
    print("ERROR: Not able to access url: " + url)
```

```python
#preview the data using pandas
import pandas as pd
import numpy as np
df = pd.read_csv(url)
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
my_plot = df.plot('Year', df.columns[1:], kind='bar', figsize=(10, 3))
my_plot.set_title('Average maximum four-hour average ozone concentrations')
_ = my_plot.legend(loc="upper left", bbox_to_anchor=(1,1))
```

```python
my_plot = df.plot('Year', df.columns[1:], figsize=(10, 4))
my_plot.set_title('Average maximum four-hour average ozone concentrations')
_ = my_plot.legend(loc="upper left", bbox_to_anchor=(1,1))
```
